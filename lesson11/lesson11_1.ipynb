```python
from pydantic import BaseModel,Field,field_validator
class Record(BaseModel):
    Company_name:str = Field(alias='Name_Ins')
    Number:str = Field(alias= 'No')
    Address:str = Field(alias= 'Address')
    Tax_number:str = Field(alias ='tax_ID_number')
    wgsX:float = Field(alias = 'wgs84aX')    
    wgsY:float = Field(alias = 'wgs84aY') 

    @field_validator("wgsX","wgsY",mode='before')  #預設是after，改成before 這邊會先做 
    @classmethod
    def val(cls, value):  # val= method name可隨意設定
        if value == '':
            return 0.0
        else:
            return value
         
class Result(BaseModel):
    records:list[Record]   #list is reocrd 的type hint
    
```

```python
import json
from pprint import pprint  #從pprint module import出pprint的function. https://docs.python.org/zh-tw/3/library/pprint.html
from pydantic import BaseModel,Field


with open('Newtaipeifood.json',encoding='utf-8') as file:
    json_content:str = file.read()
    all_data = json.loads(json_content)  #all_data是python資料結構
    ## pprint(all_data)

all_data['result']

result:Result = Result.model_validate(all_data['result'])  #model_validate 是驗證python, _json是驗證json的資料結構

```
