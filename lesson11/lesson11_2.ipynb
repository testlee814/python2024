```python
import requests
from requests import Response

aqi_url = 'https://data.moenv.gov.tw/api/v2/aqx_p_488?api_key=e8dd42e6-9b8b-43f8-991e-b3dee723a52d&limit=1000&sort=datacreationdate desc&format=JSON'
res:requests.Response = requests.get(aqi_url)  #reposnse是reuqest的class

if res.status_code == 200:
    print("Download successfully!!!")
else:
    print("Download Failed!!! ")
```

```python
from pprint import pprint
from pydantic import BaseModel, Field, field_validator
## pprint(res.text)  印出所有資料

class Site(BaseModel):
    sitename:str
    county:str
    aqi:int
    status:str
    pm25:float = Field(alias = 'pm2.5')
    so2:float 
    datacreationdate:str
 
    @field_validator("so2","pm25",mode='before')  #預設是after，改成before 這邊會先做 
    @classmethod
    def empty_to_zero(cls, value:str) -> str: # val= method name可隨意設定
        if value == '':
            return 0.0
        else:
            return value
class AQI(BaseModel):
    records:list[Site]
    
```

```python
all_data:AQI = AQI.model_validate_json(res.text)
all_data 
```

```python
for site in all_data.records:
    print(site)

```
